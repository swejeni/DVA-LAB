### Column wise and multliplefunction application

```python
tips=pd.read_csv('tips.csv')
tips
```

```python
#add tip percentage of total bill
tips['tip_pct']=tips['tip']/tips['total_bill']
tips[:6]
```

```python
grouped=tips.groupby(['day','smoker'])
grouped.count()
```

```python
grouped_pct=grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
grouped_pct.agg(['mean','std',peak_to_peak])#agg used for more than one operation#2 predefind 1 user defind
```

```python
grouped_pct.agg([('foo','mean'),('bar','std')])
```

```python
grouped_pct.agg([('foo','mean'),('bar',np.std)])
```

```python
functions=['count','mean','max']
result=grouped['tip_pct'].agg(functions)
result
```

```python
functions=['count','mean','max']
result=grouped['tip_pct','total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
ftuples=[('Durch','mean'),('abweu',np.var)]#tuples#row base
grouped['tip_pct','total_bill'].agg(ftuples)
```

```python
grouped.agg({'tip':np.max,'size':'sum'})#dictionary
```

```python
grouped.agg({'tip_pct':[min,'max','mean',np.std],'size':'sum'})#multiple values#(min and max -quotes not needed)
```

#### Apply general split-apply-combine

```python
def top(df,n=5,column='tip_pct'):#default arguments
    return df.sort_values(by=column)[-n:]
top(tips,n=6)
```

```python
tips.groupby('smoker').apply(top)#by using apply method
```

```python
tips.groupby(['smoker','day']).apply(top,n=5,column='total_bill')#sorting left to right
```

```python
tips.groupby(['smoker','day']).apply(top,n=1,column='total_bill')
```

```python
result=tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack()
```

#### suppressig the group key

```python
tips.groupby('smoker',group_keys=False).apply(top)
```
